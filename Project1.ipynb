```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("banking_data.csv")
age_bins = range(0, 101, 10)  # 0-9, 10-19, ..., 90-99, 100+
age_labels = [f"{i}-{i+9}" for i in age_bins[:-1]]
df['age_group'] = pd.cut(df['age'], bins=age_bins, right=False, labels=age_labels)
print(df['age_group'].value_counts().sort_index())
print(df.head())

sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Number of Clients')
plt.show()
```

```python
job_counts = df['job'].value_counts()
print(job_counts)

job_proportions = df['job'].value_counts(normalize=True)
print(job_proportions)

sns.countplot(data=df, x='job')
plt.title('Distribution of Job Types Among Clients')
plt.xlabel('Job Type')
plt.ylabel('Number of Clients')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("banking_data.csv")
plt.figure(figsize=(8,4))
sns.histplot(df['previous'], bins=range(0, df['previous'].max()+2), kde=False)
plt.title('Number of Contacts Before Current Campaign')
plt.xlabel('Number of Previous Contacts')
plt.ylabel('Number of Clients')
plt.xlim(0,50)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='poutcome')
plt.title('Outcomes of Previous Marketing Campaigns')
plt.xlabel('Previous Outcome')
plt.ylabel('Number of Clients')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='y')
plt.title('Term Deposit Subscription Distribution')
plt.xlabel('Subscribed (yes/no)')
plt.ylabel('Number of Clients')
plt.show()
```

```python
# Example: Correlation heatmap for numeric columns
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Numeric Features)')
plt.show()

# Example: Compare mean duration for subscribers vs. non-subscribers
print(df.groupby('y')['duration'].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataframe
df = pd.read_csv("banking_data.csv")
# Analyze the relationship between 'job' and 'y'
job_vs_y = pd.crosstab(df['job'], df['y'])
print(job_vs_y)

job_vs_y.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Term Deposit Subscription by Job Type')
plt.xlabel('Job Type')
plt.ylabel('Number of Clients')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
